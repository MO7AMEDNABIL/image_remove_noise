```python
import matplotlib.pyplot as py
import cv2 as cv
import numpy as np
```

```python
jude = cv.imread('Jude-Bellingham-1-e1692476140298.jpeg')
py.imshow(jude) #show the image 
```

```python
jude_rgb = cv.cvtColor(jude,cv.COLOR_BGR2RGB) #convert the image from BGR TO RGB
py.imshow(jude_rgb)
py.axis('off') #hide the axis
```

```python
gauss_blur = cv.GaussianBlur(jude_rgb,(7,7),6) #apply the GaussianBlur filter on img 

py.imshow(gauss_blur)
```

```python
median_blur = cv.medianBlur(jude_rgb,5) #apply the medianBlur filter on img 
py.imshow(median_blur)
```

```python
fig = py.figure(figsize=(10, 10)) #creating figure to show the two images to see the Differs between them after sharpening

fig.add_subplot(2,2,1) # add the first image at first phase

# showing image
py.imshow(jude_rgb)
py.axis('off') #hide the axis
py.title("original") # image title

# Adds a subplot at the 2nd position
fig.add_subplot(2,2,2) # add the second image at first phase

py.imshow(gauss_blur)
py.axis('off')  #hide the axis
py.title("GaussianBlur") # image title

fig.add_subplot(2,2,3) # add the second image at first phase

py.imshow(median_blur)
py.axis('off')  #hide the axis
py.title("medianBlur") # image title
```
